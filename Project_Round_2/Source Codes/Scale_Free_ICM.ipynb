```python
import pandas as pd # Data I/O
import numpy as np # Data Manipulation

import networkx as nx # Network Ananlysis

import ndlib.models.epidemics as ep # ICM Model
import ndlib.models.ModelConfig as mc # ICM Model
```

```python
#Create a 50 node scale free network

G=nx.scale_free_graph(50,seed=21)
```

```python
model_1 = ep.IndependentCascadesModel(G)

config = mc.Configuration()

threshold = 0.2
for e in G.edges():
    config.add_edge_configuration("threshold", e, threshold)

config.add_model_parameter('fraction_infected', 1/len(G.nodes()))

model_1.set_initial_status(config)

model_1.iteration_bunch(20,node_status=False)
```
